```python
 from PIL import Image
import matplotlib.pyplot as plt

image_path = "/content/image.jpg"
img = Image.open(image_path)

formats = ["JPEG", "PNG", "BMP"]

# Save images in different formats
for fmt in formats:
    output_filename = f"output.{fmt.lower()}"  # Create a variable for the output filename
    img.save(output_filename, fmt)
    print(f"Saved image as {output_filename}") # Print confirmation message

# Open and display images
for fmt in formats:
    output_filename = f"output.{fmt.lower()}"  # Use the same variable for consistency
    with Image.open(output_filename) as img:
        print(f"Format: {img.format}")
        print(f"Size: {img.size}")
        print(f"Mode: {img.mode}")
        plt.imshow(img)
        plt.title(f"Image in {img.format} format")
        plt.axis('off')
        plt.show()
        print()
```
