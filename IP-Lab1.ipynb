```python
import cv2 as cv
from matplotlib import pyplot as plt
import os
```

```python
a=os.listdir()
b=[]
for i in a:
  if i.endswith(".jpg"):
    b.append(i)
b
```

```python
#img=cv.imread('images6.jpg')
#img3=cv.imread('images2.jpg')
```

```python
for i in b:
  img=cv.imread(i)
  r,g,b=cv.split(img)
  img2=cv.merge([g,b,r])
  img3=cv.merge([b,r,g])
  plt.figure(figsize=(15,15))
  plt.subplot(131)
  plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
  plt.title("Original image")
  plt.subplot(132)
  plt.title("New image")
  plt.imshow(cv.cvtColor(img2,cv.COLOR_BGR2RGB))
  plt.subplot(133)
  plt.title("New image part 2")
  plt.imshow(cv.cvtColor(img3,cv.COLOR_BGR2RGB))
  plt.show()
```
